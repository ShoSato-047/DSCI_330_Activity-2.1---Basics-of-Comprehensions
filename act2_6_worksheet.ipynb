```python
def unfoldr(empty, func, s):
    out = []
    while not empty(s):  # Stop when string is empty, otherwise keep going
        value, s = func(s)  # Apply decomposition function
        out.append(value)
    return out

# Define the empty function (checks if the string is empty)
empty = lambda s: len(s) == 0

# Define the decompose function using partition
def func(s):
    first, sep, rest = s.partition(',')  # Splits at the first comma
    return first, rest  # Return first part and remaining strings
```

```python
# Define comma_split using unfoldr
comma_split = lambda s: unfoldr(empty, func, s)
```

```python
comma_split('a,b,c')
```
